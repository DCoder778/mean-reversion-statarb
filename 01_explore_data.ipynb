```python
import matplotlib.pyplot as plt
import yfinance as yf
# Imports yfinance to download stock data, and matplotlib to plot it


```

```python
df = yf.download("KO", start="2020-01-01", end="2025-01-01")
# Downloads Coca-Cola's stock data using its ticker, for 2020 to 2025
```

```python
df.head(15)
# Displays the first 15 rows of the dataframe
```

```python
df['Close'].plot(figsize=(10, 5), title="Coca Cola Stock Closing Price")
plt.show()
# Plots the closing price of Coca-Cola stock over time
```

```python
returns = df["Close"].pct_change()
# Computes the daily % change in closing price
```

```python
returns.head(10)
# Displays the first 10 daily return values
```

```python
returns.plot(kind="hist", bins=100, figsize=(10, 5), title="Coca Cola Daily Returns Distribution")
plt.show()
# Plots a histogram showing the distribution of daily returns
```

```python
df['Close'].plot(figsize=(10, 5), label="Closing Price")
df['Close'].rolling(15).mean().plot(label="15-Day Rolling Average")
# Plots the 15-day rolling average of Coca-Cola's closing price
plt.legend()
plt.title("Coca Cola Closing Price vs 15-Day Rolling Average")
plt.show()
```

```python
df['Close'].plot(figsize=(10, 5), label="Closing Price")
df['Close'].rolling(20).mean().plot(label="20-Day Rolling Average")
df['Close'].rolling(50).mean().plot(label="50-Day Rolling Average")
# Plots the 20-day and 50-day rolling averages of Coca-Cola's closing price
plt.legend()
plt.title("Coca Cola Closing Price vs 20-Day and 50-Day Rolling Average")
plt.show()
```
