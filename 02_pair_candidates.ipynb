```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df_a = yf.download('KO', start = '2020-01-01', end = '2025-01-01')
df_b = yf.download('PEP', start = '2020-01-01', end = '2025-01-01')
norm_a = df_a['Close']/df_a['Close'].iloc[0]
norm_b = df_b['Close']/df_b['Close'].iloc[0]
norm_a.plot(label = 'KO', figsize = (10,5))
plt.title('Normalised price of Coca-Cola')
norm_b.plot(label = 'PEP', figsize = (10,5))
plt.title('Normalised price of Pepsi')
plt.legend()
plt.show()

ratio = df_a['Close']['KO']/df_b['Close']['PEP']
ratio.plot(figsize = (10,5), title = "Price Ratio of Coca-Cola and Pepsi")





```

```python
df_c = yf.download('MCD', start = '2020-01-01', end = '2025-01-01')

df_d = yf.download('YUM', start = '2020-01-01', end = '2025-01-01')

norm_c = df_c['Close']/df_c['Close'].iloc[0]

norm_d = df_d['Close']/df_d['Close'].iloc[0]

norm_c.plot(label = 'MCD', figsize = (10,5))
plt.title('Normalised Price of McDonalds')

norm_d.plot(label = 'YUM', figsize = (10,5))
plt.title('Normalised Price of Yum Brands')
plt.legend()

plt.show()

ratio = df_c['Close']['MCD']/df_d['Close']['YUM']
ratio.plot(figsize = (10,5), title = "Price Ratio of McDonalds and Yum Brands")
plt.show()
```
